## Generated with Microsoft Copilot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('study_performance.csv')

# Prepare the data
scores = df[['math_score', 'reading_score', 'writing_score', 'gender']]

# Melt the DataFrame to make it suitable for boxplot
scores_melted = pd.melt(scores, id_vars='gender', var_name='score_type', value_name='score')

# Create the boxplot
plt.figure(figsize=(10, 6))
scores_melted.boxplot(column='score', by=['score_type', 'gender'], grid=False)
plt.title('Scores by Gender')
plt.suptitle('')  # Suppress the automatic title
plt.xlabel('Score Type and Gender')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('study_performance.csv')

# Prepare the data
scores = df[['math_score', 'reading_score', 'writing_score', 'gender']]

# Melt the DataFrame to make it suitable for boxplot
scores_melted = pd.melt(scores, id_vars='gender', var_name='score_type', value_name='score')

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='score_type', y='score', hue='gender', data=scores_melted, palette='Set3')
plt.title('Scores by Gender')
plt.xlabel('Score Type')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

## Generated with Gemini


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Path to your CSV file
csv_file_path = "study_performance.csv"

# Read the CSV data into a Pandas DataFrame
data = pd.read_csv(csv_file_path)

# Create the boxplot using seaborn
sns.boxplot(
    x = "gender",
    y = ["math_score", "reading_score", "writing_score"],
    data=data
)

# Customize the plot (optional)
plt.xlabel("Gender")
plt.ylabel("Score")
plt.title("Distribution of Scores by Gender")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
